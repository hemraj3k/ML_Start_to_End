We will now make a distribution plot of the math score of the students. A distribution plot tells us how the data is distributed. We will use the distplot function.

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Sample dataset
import pandas as pd
import numpy as np
df = pd.DataFrame({'math score': np.random.randint(40, 100, 100)})

# Plot histogram with KDE (Kernel Density Estimation)
sns.histplot(df['math score'], bins=20, kde=True, color='blue')

# Formatting
plt.title("Distribution of Math Scores")
plt.xlabel("Math Score")
plt.ylabel("Frequency")
plt.show()
```

We will now look at the correlation between the 3 scores with the help of a heatmap. For this, we will use corr() and heatmap() function for this exercise.

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Load dataset
df = pd.read_csv('StudentsPerformance.csv')

# Convert categorical columns to numeric using One-Hot Encoding
df_encoded = pd.get_dummies(df, drop_first=True)

# Compute correlation matrix
corr = df_encoded.corr()

# Plot heatmap
plt.figure(figsize=(10, 6))  # Adjust figure size
sns.heatmap(corr, annot=True, cmap="coolwarm", fmt=".2f", square=True)

# Adjust labels
plt.yticks(rotation=0)
plt.xticks(rotation=45)
plt.title("Feature Correlation Heatmap")

plt.show()
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Load dataset
df = pd.read_csv('StudentsPerformance.csv')

# Create scatter plot with gender-based coloring
sns.relplot(x='math score', y='writing score', hue='gender', data=df, aspect=1.5)

# Formatting
plt.title("Math Score vs Writing Score by Gender")
plt.xlabel("Math Score")
plt.ylabel("Writing Score")

plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load dataset
df = pd.read_csv('StudentsPerformance.csv')

# Compute mean scores by parental education level
score_means = df.groupby('parental level of education')[['math score', 'reading score', 'writing score']].mean().T

# Plot the transposed data
plt.figure(figsize=(12, 8))
score_means.plot(kind='bar', figsize=(12, 6), colormap='viridis')

# Formatting
plt.title("Average Scores by Parental Education Level")
plt.xlabel("Subjects")
plt.ylabel("Average Score")
plt.legend(title="Parental Education Level", bbox_to_anchor=(1.05, 1), loc='upper left')
plt.xticks(rotation=0)  # Keep labels readable

plt.show()
```
