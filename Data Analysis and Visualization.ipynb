```python
import matplotlib.pyplot as plt
import numpy as np
# Data for plots
x = np.linspace(0, 10, 100)
y = np.sin(x)
# Line plot
plt.subplot(2, 2, 1)
plt.plot(x, y)
plt.title('Line Plot')
# Scatter plot
plt.subplot(2, 2, 2)
plt.scatter(x, y)
plt.title('Scatter Plot')
# Bar chart
plt.subplot(2, 2, 3)
plt.bar([1, 2, 3], [10, 20, 30])
plt.title("Bar Chart")
# Histogram
plt.subplot(2, 2, 4)
plt.hist(y, bins=10)
plt.title('Histogram')
plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Sample dataset
df = pd.read_csv("data.csv")  # Load your dataset

# Convert categorical columns to numeric
df_encoded = pd.get_dummies(df, drop_first=True)  # One-Hot Encoding

# Generate heatmap
plt.figure(figsize=(8, 6))  # Set figure size
sns.heatmap(df_encoded.corr(), annot=True, cmap="coolwarm", fmt=".2f")
plt.title("Feature Correlation Heatmap")
plt.show()
```

```python
sns.boxplot(x=df['Salary'])
plt.title("Outliers in Salary")
plt.show()
```

```python
Q1 = df['Salary'].quantile(0.25)
Q3 = df['Salary'].quantile(0.75)
IQR = Q3 - Q1

# Define lower and upper bounds
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

# Remove outliers
df_clean = df[(df['Salary'] > lower_bound) & (df['Salary'] < upper_bound)]
print(df_clean)
```
